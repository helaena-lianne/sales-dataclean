## Data Cleaning

```python
## import statements

import os
import re
import PIL
import numpy as np
import pandas as pd
import seaborn as sns
from joblib import dump, load
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df = pd.read_csv("/content/drive/MyDrive/datasets/Sales_Dataset_2024.csv")
display(df)
```

```python
df.info()
```

drop duplicate rows

```python
print(f"Number of duplicates: {df.duplicated().sum()}")
df.drop_duplicates(inplace=True)
print(f"Number of duplicates after: {df.duplicated().sum()}")
```

displays the number of null values

```python
df.isnull().sum()
df.isna().sum()
```

drop the rows with null values

```python
df = df.dropna()
```

converts the date column to a datetime datatype then drops the time portion

```python
df["Date"] = pd.to_datetime(df["Date"])
df["Date"] = df["Date"].dt.date
```

displays the unique values of the Region column

```python
df["Region"].unique()
```

corrects the mispelled values and corrects the cases for uniformity

```python
df["Region"] = df["Region"].replace({
    "westt": "West",
    "Easst": "East",
    "NORTH": "North",
    "north":"North",
    "south":"South"
})
```

```python
df["Region"].unique()
```

displays the unique values of the Product column

```python
df["Product"].unique()
```

```python
df["Product"] = df["Product"].replace({
    "tabllet": "Tablet",
    "MOBLIE": "Mobile",
    "SMARTWATCH": "Smartwatch",
    "headPhones":"Headphones",
    "laptop":"Laptop"
})
```

```python
df["Product"].unique()
```

```python
df["Salesperson"].unique()
```

```python
df["Category"].unique()
```

converts the dataframe to csv

```python
df.to_csv('sales_data.csv', index=False)
print("DataFrame saved to sales_data.csv")
```
